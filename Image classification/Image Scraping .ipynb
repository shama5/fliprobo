# Part 1 - Image Collection

```python
import selenium
import pandas as pd
from selenium import webdriver
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
url='https://www.amazon.in/'
```

```python
# Function to search 300 images
def extract_image(x):
    #  Enter x in “Search” field 
    
    l=0 # To keep count of number of images (upto 300)
    
    driver=webdriver.Chrome("chromedriver.exe")
    driver.get(url)
    
    search=driver.find_element_by_id("twotabsearchtextbox")
    search.send_keys(x)

    # click the search button

    search_buton=driver.find_element_by_id("nav-search-submit-button")
    search_buton.click()
    time.sleep(5)
    
    images=[]
    
    while (l<=300):
        elements=driver.find_elements_by_xpath("//img[@class ='s-image' ]")
        l=l+len(elements)
        
        for i in elements:
            image=i.get_attribute('src')
            images.append(image)
        
        if (l<=300):                    
            # Clicking on next page
            driver.find_element_by_xpath("//li[@class='a-last']/a").click()
            time.sleep(5)
    driver.close()       
    return(images)
            
```

```python
saree=extract_image('Sarees (women)')
len(saree)
```

```python
trouser=extract_image('Trousers (men)')
len(trouser)
```

```python
jeans=extract_image('Jeans (men)')
len(jeans)
```

```python
# Creating Directories
import os


# function to create a directory

def directory(folder):
    img_dir = os.path.join(os.getcwd(), folder)
    if not os.path.exists(img_dir):
        os.makedirs(img_dir)
        
    return folder

# Calling the above function

folder = directory("Image")
        
directory('Image/Sarees_Images')
directory('Image/Trousers_Images')
directory('Image/Jeans_Images')

# Dowloading images

import shutil
import requests
```

```python
for index, link in enumerate(saree):
    print('Downloading {0} of total saree images'.format(index+1))
    response=requests.get(link)
    with open('Image/Sarees_Images/img{0}.jpeg'.format(index+1),"wb") as file:
        file.write(response.content)
        
```
